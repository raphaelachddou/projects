```python
feat_df = pd.read_csv('data/features.csv', sep=',')
```

```python
def feature_color_meanRed(idRep):
    t = time.time()
    m=0
    for x in idRep:
        m+=1
        filename = 'data/im_resized/{}_resized.jpg'.format(x)
        filename_S = 'data/im_resized/{}_Segmentation_resized.jpg'.format(x)
        image = imread(filename)
        image_S = imread(filename_S)
        image_S = image_S.astype(bool)
        
        #toutes les valeurs hors de la lésion doivent être nulles
        masked_im = np.array(image[:,:,0]*image_S, np.int32)
        area = np.count_nonzero(image_S )
        res = np.sum(masked_im)/area
        if(m%50 ==0):
            print(time.time()-t)
            print(m," : tour 1")
        feat_df.loc[feat_df['ImageId'] == x, 'Mean Red'] = res
    print(time.time()-t)
    feat_df.to_csv('data/features.csv', index=None, sep=',', mode='w')

    
def feature_color_meanGreen(idRep):
    t = time.time()
    m=0
    for x in idRep:
        m+=1
        filename = 'data/im_resized/{}_resized.jpg'.format(x)
        filename_S = 'data/im_resized/{}_Segmentation_resized.jpg'.format(x)
        image = imread(filename)
        image_S = imread(filename_S)
        image_S = image_S.astype(bool)
        
        #toutes les valeurs hors de la lésion doivent être nulles
        masked_im = np.array(image[:,:,1]*image_S, np.int32)
        area = np.count_nonzero(image_S )
        res = np.sum(masked_im)/area
        if(m%50 ==0):
            print(time.time()-t)
            print(m," : tour 1")
        feat_df.loc[feat_df['ImageId'] == x, 'Mean Green'] = res
    print(time.time()-t)
    feat_df.to_csv('data/features.csv', index=None, sep=',', mode='w')
    
def feature_color_meanBlue(idRep):
    t = time.time()
    m=0
    for x in idRep:
        m+=1
        filename = 'data/im_resized/{}_resized.jpg'.format(x)
        filename_S = 'data/im_resized/{}_Segmentation_resized.jpg'.format(x)
        image = imread(filename)
        image_S = imread(filename_S)
        image_S = image_S.astype(bool)
        
        #toutes les valeurs hors de la lésion doivent être nulles
        masked_im = np.array(image[:,:,2]*image_S, np.int32)
        area = np.count_nonzero(image_S )
        res = np.sum(masked_im)/area
        if(m%50 ==0):
            print(time.time()-t)
            print(m," : tour 1")
        feat_df.loc[feat_df['ImageId'] == x, 'Mean Blue'] = res
    print(time.time()-t)
    feat_df.to_csv('data/features.csv', index=None, sep=',', mode='w')
    


feature_color_meanRed(feat_df['ImageId'].values)
feature_color_meanGreen(feat_df['ImageId'].values)
feature_color_meanBlue(feat_df['ImageId'].values)
```

```python
def feature_color_stdRed(idRep):
    t = time.time()
    m=0
    for x in idRep:
        m+=1
        filename = 'data/im_resized/{}_resized.jpg'.format(x)
        filename_S = 'data/im_resized/{}_Segmentation_resized.jpg'.format(x)
        image = imread(filename)
        image_S = imread(filename_S)
        image_S = image_S.astype(bool)
        
        #toutes les valeurs hors de la lésion doivent être nulles
        masked_im = np.array(image[:,:,0]*image_S, np.int32)
        area = np.count_nonzero(image_S )
        mean = np.sum(masked_im)/area
        (h,w) = np.shape(masked_im)
        std = 0
        for i in range(h):
            for j in range(w):
                if (image_S[i,j]):
                    std +=(mean-masked_im[i,j])**2
        res = np.sqrt(std/(area))
        if(m%50 ==0):
            print(time.time()-t)
            print(m," : tour 1")
        feat_df.loc[feat_df['ImageId'] == x, 'Std Red'] = res
    print(time.time()-t)
    feat_df.to_csv('data/features.csv', index=None, sep=',', mode='w')
feature_color_stdRed(feat_df['ImageId'].values)
```
